## read all anchors

```python
import numpy as np
```

```python
ahrefs = list(soup.find_all("a"))
```

```python
print(ahrefs[1]['href'], ahrefs[1].text)
```

```python
valid_ahrefs = []
invalid_ahrefs = []
def extract_ahref(ahref):
  extracted = dict()
  extracted['tag'] = ahref
  
  # some ahref tags do not have href at all
  if 'href' in ahref.attrs and (
    # AND some ahref tags are js scripts so look for href values starting with '/' or '#' or 'http(s)'
    ahref['href'].find('/')==0 or ahref['href'].find('#')>=0 or ahref['href'].find('http')>=0
  ):
    extracted['anchor'] = ahref.text
    extracted['href'] = ahref['href']
    valid_ahrefs.append(extracted)
  
  else: invalid_ahrefs.append(extracted)

for ahref in ahrefs:
  extract_ahref(ahref)
```
